#### Data Integrity

Data in dbs must be accurate & consistant & must not be changed unauthorized. There are 3 methods to keep data integrity:

1. Constraints - Constraints are the rules or conditions to be met before any CRUD ops on data. They are done using DDL commands.
2. Transactions - Dbs operation that are treated as single unit like fund transfer in banks.
3. Normalization - Technique that minimizes data redundancy & ensures data consistancy by organizing data into seperate tables.

### Constraints in MYSQL

1. NOT NULL: where presence of value is important like train tkt data & PNR attribute (column) must be NOT NULL
2. UNIQUE: where value must be unique like email of pax must be unique or roll no of students
3. PRIMARY KEY
4. AUTO INCREMENT
5. CHECK: Its a condition which you put on a column like driving license application data & applicant AGE column, condition is that value must be > 18
6. DEFAULT: when you asign a default data like registration data which will be set default to todays date
7. FOREIGN KEY: FK helps in creating references between 2 tables

### Refrential Actions

3. SET NULL - If you delete in 1 table then any refrence in another table will be set to null
